## Calculating explicit Utility for our dishes with example of a weightloss diet

```python
def calculate_explicit_utility_weightloss(proteins_rank, minerals_rank, vitamins_rank, carbohydrates_rank, fats_rank, fibers_rank, sugars_rank):
    # Define weights for each nutrient category
    weights = {'Proteins': 3, 'Minerals': 2, 'Vitamins': 2, 'Carbohydrates': 1, 'Fats': 1, 'Fibers': 3, 'Sugars': 1}
    
    # Define mapping of rankings to weights
    ranking_weights = {'Low': 1, 'Medium': 2, 'High': 3}

    calorie=400

    # Calculate scores for each nutrient category
    scores = {'Proteins': weights['Proteins']-ranking_weights[proteins_rank] >0,
              'Minerals': (ranking_weights[minerals_rank] - weights['Minerals']) if -ranking_weights[minerals_rank] + weights['Minerals'] > 0 else 1.2*(ranking_weights[minerals_rank] - weights['Minerals']),
              'Vitamins': (-ranking_weights[vitamins_rank] + weights['Vitamins']) if -ranking_weights[vitamins_rank] + weights['Vitamins'] > 0 else 0.8*(-ranking_weights[vitamins_rank] + weights['Vitamins']),
              'Carbohydrates': (-ranking_weights[carbohydrates_rank] + weights['Carbohydrates']) if -ranking_weights[carbohydrates_rank] + weights['Carbohydrates'] > 0 else (-ranking_weights[carbohydrates_rank] + weights['Carbohydrates'])**2,
              'Fats': (-ranking_weights[fats_rank] + weights['Fats']) if -ranking_weights[fats_rank] + weights['Fats'] > 0 else (-ranking_weights[fats_rank] + weights['Fats'])**2,
              'Fibers': (-ranking_weights[fibers_rank] + weights['Fibers']),
              'Sugars': (-ranking_weights[sugars_rank] + weights['Sugars']) if -ranking_weights[sugars_rank] + weights['Sugars'] > 0 else (-ranking_weights[sugars_rank] + weights['Sugars'])**2}
    # Calculate overall utility score
    utility_score = sum(scores.values())
    
    return utility_score

explicit_dietry_utility=[]

for i in range(0, len(df)):
    print(df.iloc[i]['Proteins'], df.iloc[i]['Minerals'], df.iloc[i]['Vitamins'], df.iloc[i]['Carbohydrates'], df.iloc[i]['Fats'], df.iloc[i]['Fibers'], df.iloc[i]['Sugars'])
    utility_score = calculate_explicit_utility_weightloss(df.iloc[i]['Proteins'], df.iloc[i]['Minerals'], df.iloc[i]['Vitamins'], df.iloc[i]['Carbohydrates'], df.iloc[i]['Fats'], df.iloc[i]['Fibers'], df.iloc[i]['Sugars'])
    print("Utility Score:", utility_score)
    explicit_dietry_utility.append(utility_score)
    print("\n")

df2['Explicit Dietry Utility'] = explicit_dietry_utility
```

```python
from numpy.random import randint
from numpy.random import rand


```

```python
def decode(bounds, n_bits, bitstring):
    decoded = []
    largest = 2 ** n_bits
    total_value = 0  # Initialize the total value of the decoded variables
    for i in range(len(bounds)):
        start, end = i * n_bits, i * n_bits + n_bits
        substring = bitstring[start:end]
        chars = "".join([str(s) for s in substring])
        integer = int(chars, 2)
        value = bounds[i][0] + (integer / largest) * (bounds[i][1] - bounds[i][0])
        total_value += value  # Accumulate the decoded value
        decoded.append(value)
        
    # Scale the decoded values to ensure the sum is equal to 1
    decoded_sum = sum(decoded)
    decoded = [v / decoded_sum for v in decoded]
    
    return decoded
```

```python
def selection(pop, scores, k=3):
    # Randomly select one index from the population as the initial selection
    selection_ix = np.random.randint(len(pop))
    # Perform a tournament among k randomly selected individuals
    for ix in np.random.randint(0, len(pop), k-1):
        # Check if the current individual has a better score than the selected one
        if scores[ix] < scores[selection_ix]:
            # Update the selected individual if a better one is found
            selection_ix = ix
    # Return the best individual from the tournament
    return pop[selection_ix]
```

```python
def crossover(p1, p2, r_cross):
    # Children are copies of the parents by default
    c1, c2 = p1.copy(), p2.copy()
    # Check if recombination should occur
    if rand() < r_cross:
        # Select a crossover point (not at the end of the string)
        pt1, pt2 = randint(1, len(p1)-2), randint(1, len(p1)-2)
        # Ensure pt2 is greater than pt1
        pt1, pt2 = min(pt1, pt2), max(pt1, pt2)
        # Perform crossover in the children
        c1 = p1[:pt1] + p2[pt1:pt2] + p1[pt2:]
        c2 = p2[:pt1] + p1[pt1:pt2] + p2[pt2:]
    return [c1, c2]
     
```

```python
import random
def mutation(bitstring, r_mut):
    rand = random.random
    for i in range(len(bitstring)):
        # Check for a mutation
        if rand() < r_mut:
            # Flip the bit
            bitstring[i] = 1 - bitstring[i]
    return bitstring
```

```python
def genetic_algorithm(objective, bounds, n_bits, n_iter, n_pop, r_cross, r_mut):

    pop = [randint(0, 2, n_bits * len(bounds)).tolist() for _ in range(n_pop)]

    # track the best solution found so far
    best, best_eval = 0, objective(decode(bounds, n_bits, pop[0]))

    # iterate over generations
    for gen in range(n_iter):
        # decode the population
        decoded = [decode(bounds, n_bits, p) for p in pop]
        # normalising the decoded values to a sum of 1
        
    

        # evaluate all candidates in the population
        scores = [objective(d) for d in decoded]
        # check for a new best solution
        for i in range(n_pop):
            if scores[i] < best_eval:
                best, best_eval = pop[i], scores[i]
                print(">%d, new best f(%s) = %f" % (gen,  decoded[i], scores[i]))

        # select parents
        selected = [selection(pop, scores) for _ in range(n_pop)]

        # create the next generation - children
        children = list()
        for i in range(0, n_pop, 2):
            # get selected parents in pairs
            p1, p2 = selected[i], selected[i + 1]
            # crossover and mutation
            for c in crossover(p1, p2, r_cross):
                # mutation
                mutation(c, r_mut)
                # store for next generation
                children.append(c)
        # replace the population
        pop = children
    return [best, best_eval]
```

```python
# define range for input
bounds = [[0, 0.5], [0, 0.5]]
# define the total iterations
n_iter = 100
# bits per variable
n_bits = 8
# define the population size
n_pop = 100
# crossover rate
r_cross = 0.9
# mutation rate
r_mut = 1.0 / (float(n_bits) * len(bounds))
# perform the genetic algorithm search
best, score = genetic_algorithm(objective, bounds, n_bits, n_iter, n_pop, r_cross, r_mut)
print('###########################################################')
decoded = decode(bounds, n_bits, best)
print('The result is (%s) with a score of %f' % (decoded, score))
```
